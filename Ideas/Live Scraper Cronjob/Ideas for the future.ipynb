# Scraping CricInfo Data: Cronjob for the future

### Motivation

Cricinfo provides detailed analytics (like pitch heatmaps for each player, player vs player comparison etc) for each map in the JSON format. However, these analytics are only accessible while the match is in progress. It would be a good idea to run a cron job on an EC2 instance or a Google App Engine app to scrape these everyday.

These data sources were discovered by tracing calls from the HTML page of a live cricket match on cricinfo.

```python
import requests
from IPython.core.display import Image
```

### Getting overall data

This is what the data on a live cricinfo page looks like. Further details can be seen upon playing around with it.

```python
Image('overview.png')
```

```python
r = requests.get('http://www.espncricinfo.com/ci/engine/match/live/centre/750675.json?innings=1')
r.json()
```

### Getting batsmen's data

This is an example of what a batsman's data looks like.

```python
Image('batsman.png')
```

```python
r = requests.get('http://www.espncricinfo.com/ci/engine/match/live/centre/750675.json?batsman=7419;card=batting;innings=1')
r.json()
```

### Getting bowler's data

This is an example of how a bowler's data looks like.

```python
Image('bowler.png')
```

```python
r = requests.get('http://www.espncricinfo.com/ci/engine/match/live/centre/750675.json?bowler=46597;card=bowling;innings=1')
```

```python
r.json()
```
